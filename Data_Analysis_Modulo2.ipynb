```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
#df.head()
```

# **Intenficando y manipulando los valores perdidos.**

```python
import numpy as np

# reemplazando los valores "?" a NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```

Evaluando los valores perdidos como Verdadero o Falso.
Utilizar las siguientes funciones.
.isnull(). 

.notnull()

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
# para contar los valores verdaderos o falsos en cada columna del data frame, se utiliza un ciclo for.
for column in missing_data.columns.values.tolist():
  for column in missing_data.columns.values.tolist(): 
    print(column)
    print (missing_data[column].value_counts())
    print("")    
```

En resumen se observa que tenemos los siguientes valores nulos.

1.- "normalized-losses": 41 missing data -- Reemplazar por Media de valores.

2.-"num-of-doors":        2 missing data.-->Reemplazar con el valor 4. Porque los autos de 4 puertas representas el 85% de los autos.

3.-"bore":                4 missing data-->Reemplazar por Media de valores.

4.-"stroke" :             4 missing data-->Reemplazar por Media de valores.

5.-"horsepower":          2 missing data-->Reemplazar por Media de valores.

6.-"peak-rpm":            2 missing data-->Reemplazar por Media de valores.

7.-"price":               4 missing data--> Borrar las 4 columnas porque es el valor que deseamos predecir. Un dato de entrada sin valor en precio no puede ser usado para la predicción, por la tanto cualquier fila sin valor en precio no es útil para nosotros.


```python
#calcular la media de la columna "normalized-losses" y reemplazar los valores "NaN" 
avg_norm_loss=df["normalized-losses"].astype("float").mean(axis=0)
print("Promedio de valores Normalized-losses",avg_norm_loss)
```

```python
#reemplazar los valores Nan
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
#Calular la media para la columna "bore" y reemplazar los valores perdidos por la media. 
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
#reemplanzando por la media.
df["bore"].replace(np.nan, avg_bore, inplace=True)
```
